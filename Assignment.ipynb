<h2> Step 2: Data Cleaning and Data Preprocessing </h2>

<p> In previous step we see find there are columns are completely null so lets find the no of columns are totally null using isna().all() find the columns are totally null are not are .sum() method sum up the all columns so according to out put there are 108 columns are there which completely null. those columns cannot use for analysis. So those columns need to remove</p>

```python
# Count the number of columns that are completely null
null_columns_count = df.isna().all().sum()

print(null_columns_count)
```

the .dropna() method is use to remove the columns or row which have null values we can set options according to aur need like, axis which can be set 1 for columns and 0 for rows. and the other option is how which can set any which need atleast one cell to be null to remove and all which need every cell to null for remove column or row . In below code axis set to 1 and how set to be all which drops columns which as all nill values after removing columns generates new dataframe which as 33 columns that stored in df_cleaned dataframe

```python
# Drop columns that are completely null
df_cleaned = df.dropna(axis=1, how='all')

print(df_cleaned)
```

In the given dataframe some  column as only one unique valuel, so it can be drop from the data because doesn't have any use

```python
# Identify columns with only one unique value
unique_counts = df_cleaned.nunique()

# Display columns with only one unique value
print("Columns with only one unique value:\n", unique_counts[unique_counts == 1].index)
```

```python
# Drop columns with only one unique value
df_cleaned = df_cleaned.drop(columns=unique_counts[unique_counts == 1].index)

# Display the cleaned DataFrame
print("\nDataFrame after dropping columns with only one unique value:\n", df_cleaned)
```

Since there are three columns with null values two of them are having low percent of null values which can handle by replace the null values with mean, median or some other imputations. 

The three columns runtime and fuel_level has very low percentage of missing values so that can be handle with mean or median imputations in this case median imputation best option because it Robustness to Outliers and Central Tendency one imporatnt note is that this two column doesn't contain any zeros this only contains NA values so direct imputaion can be done.

```python
df_cleaned['runtime'].fillna(df_cleaned['runtime'].median(), inplace=True) #replaceing null values with median for runtime column

df_cleaned['fuel_level'].fillna(df_cleaned['fuel_level'].median(), inplace=True) #replaceing null values with median for fuel_level column
 df_cleaned['adblue_level'].fillna(df_cleaned['adblue_level'].median(), inplace=True) #replaceing null values with median for fuel_level column
```

from above box there are outliers present in thr rpm columns so to remove there are two method remove the outliers or capping the outlier methodlets perform both fo both method IQR use which use to lower limit and upper limit

```python
percentile25 = df['rpm'].quantile(0.25) #1st quartile of the column 
percentile75 = df['rpm'].quantile(0.75) #3rd quartile of the column
percentile75
iqr = percentile75 - percentile25 #calculating IQR 
```

```python
upper_limit = percentile75 + 1.5 * iqr #finding upper limit and lower limit which consist 3 standard deviation which will over 90 percentile
lower_limit = percentile25 - 1.5 * iqr
print(upper_limit)
lower_limit
```

finding the outliers uisng lower and upper limit

```python
df_cleaned[df_cleaned['rpm'] > upper_limit]
```

```python
new_df = df_cleaned[(df_cleaned['rpm'] < upper_limit) & (df_cleaned['rpm'] > lower_limit)]
```
